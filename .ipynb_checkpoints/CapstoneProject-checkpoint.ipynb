# Data Processing

#### Now we can move onto data processsing; which will be comprised of transformation, one-hot encoding, and test/train dataset split. 

```python
# We need to log transform the 'loss' variable
from math import log1p

#print('Before')
sns.pairplot(train_data[cont_columns], vars=['loss'], kind = 'scatter',diag_kind='kde') #before the transformation
print('\n')

#print('After')
train_data['loss'] = train_data['loss'].apply(log1p) #log transform the loss variable
sns.pairplot(train_data[cont_columns], vars=['loss'], kind = 'scatter',diag_kind='kde') #plot the newly transformed loss
```

```python
# One-hot encoding all the categorical data 
# as mentioned by another kaggler/data analyst, the cat data in both train and test data do not contain the same amount of attributes
# therefore its best for us to initially create a union set of both categorical data in test/train dataset

#Import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

#Init an empty to list to store all the labels
labels = []

for i in range(116): #116 b/c we have 116 categorical columns
    train = train_data[cat_columns[i]].unique() #
    test = test_data[cat_columns[i]].unique()
    labels.append(list(set(train) | set(test)))    

#One hot encode all categorical attributes
cats = []
for i in range(116):
    #Label encode
    label_encoder = LabelEncoder()
    label_encoder.fit(labels[i])
    feature = label_encoder.transform(train_data.iloc[:,i])
    feature = feature.reshape(train_data.shape[0], 1)
    #One hot encode
    onehot_encoder = OneHotEncoder(sparse=False,n_values=len(labels[i]))
    feature = onehot_encoder.fit_transform(feature)
    cats.append(feature)

# Make a 2D array from a list of 1D arrays
encoded_cats = np.column_stack(cats)

#Concatenate encoded attributes with continuous attributes
dataset_encoded = np.concatenate((encoded_cats,train_data[cont_columns]),axis=1)

print(dataset_encoded.shape)
```

```python
# We can now split data into train and validation

#import the scikit train_test_split library
from sklearn import model_selection

# init the seed for reproducability of the experiment
# any number would suffice
seed = 42

# init the X,Y as our inputs into train_test_split
# we slice the newly created 'dataset_encoded' into X,Y 
X = dataset_encoded[:,0:1190] # we choose all rows, all columns except the last one as our choice for X
Y = dataset_encoded[:,1190] # we choose all rows, only the very last column as our choice for Y

# now we can create the train,test splts 
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, Y, test_size = 0.20, random_state=seed)


#print(X_train)
#print(X_test)
#print(y_train)
#print(y_test)
```

# Model Evaluation

#### Now we can start implementing the various algorithms, evaluating, and finally analyzing the subsequent prediction quality

```python
# Instantiate the metric evalaution model

#Metric evaluation parameter
from sklearn.metrics import mean_absolute_error

# instantiate a list where we'll store all the scores from various algorithms for visualizing them later
Score_list = []
algo_names = []
```

```python
# transform back the Test set since we normalized it earlier
# this has to be done before we use the metric model
y_test = np.expm1(y_test)
X_test = np.expm1(X_test)
```

```python
# First we're going to implement a simple linear regression

# import the algorithm
from sklearn.linear_model import LinearRegression

# instantiate the algo
LR = LinearRegression(n_jobs=-1)

# liear regression score list
# we're going to take avergae since we're adding the MAE score for each column separately 
scores = []

for i in range(1,1190):
    
    #fit the model 
    LR.fit(X_train[:,i].reshape(-1, 1), y_train)

    # our model prediction
    prediction = LR.predict(X_test[:,i].reshape(-1, 1))
    
    # obtain the result
    lr_result = mean_absolute_error((y_test), (prediction))
    
    # add the score to Score_list
    scores.append(lr_result)
    
Score_list.append(np.sum(scores)/1191)
algo_names.append("LR")
```

```python
print("The MAE score for the Linearegression model above is: %d" % lr_result)
```

```python
# Second we'll try Decision Tree Regressor

# start off by importing the model
from sklearn.tree import DecisionTreeRegressor

# instaniate the algo
DTR = DecisionTreeRegressor(random_state=seed)

# fit the model
DTR.fit(X_train,y_train)

# our model prediction
DTR_prediction = DTR.predict(X_test)

# obtain the result
dtr_result = mean_absolute_error((y_test), (DTR_prediction))

#add the score to Score_list
Score_list.append(dtr_result)
algo_names.append("DTR")
```

```python
print("The MAE score for the DecisionTreeRegression model above is %d" % (dtr_result))
```

```python
# third we'll try SGDregression

# import the model
from sklearn.linear_model import SGDRegressor

# instantiate the model
SGD = SGDRegressor(random_state=seed,max_iter=100000,tol=1e-3,epsilon=0.01,alpha=0.01)

# fit the model
SGD.fit(X_train,y_train)

# our model prediction
SGD_prediction = SGD.predict(X_test)

# obtain the mae score
SGD_result = mean_absolute_error((y_test), (SGD_prediction))

# add the score to score_list
Score_list.append(SGD_result)
algo_names.append("SGD")
```

```python
print("The MAE score for the SGD model above is %d" % (SGD_result))
```

```python
# fourth we'll try RandomForestRegressor

# import the model
from sklearn.ensemble import RandomForestRegressor

# instantiate the model
RFR = RandomForestRegressor(random_state=seed,n_jobs=-1,criterion='mse') #Please do not use 'mas' as criterion due to a bug that cuases the process to go on forever 

# fit the model
RFR.fit(X_train,y_train)

# our model's prediction
RFR_prediction = RFR.predict(X_test)

# obtain the mae result
RFR_result = mean_absolute_error((y_test), (RFR_prediction))

# add the score to the score_list
Score_list.append(RFR_result)
algo_names.append("RFR")
```

```python
print("The MAE score for the RFR model above is %d " % (RFR_result))
```

```python
# Fifth we'll try GradientBoostingRegressor

# import the model
from sklearn.ensemble import GradientBoostingRegressor

# instantiate the model
GBR = GradientBoostingRegressor(random_state=seed, n_estimators=25,verbose=True) #Please do not use 'mas' as criterion due to a bug that cuases the process to go on forever

# fit the model
GBR.fit(X_train, y_train)

# our model's prediction
GBR_prediction = GBR.predict(X_test)

# obtain the mae result
GBR_result = mean_absolute_error((y_test), (GBR_prediction))

# add the score to the score_list
Score_list.append(GBR_result)
algo_names.append("GBR")
```

```python
print("The MAE score for the GBR model above is %d " % (GBR_result))
```

```python
# Finaly we'll use the MLP regressor

# import the model
from sklearn.neural_network import MLPRegressor

# instantiate the model
MLP = MLPRegressor(random_state=seed, max_iter=30)

# fit the model
MLP.fit(X_train, y_train)

# our models prediction
MLP_prediction = MLP.predict(X_test)

# obtain the mae result
MLP_result = mean_absolute_error((y_test), (MLP_prediction))

# add the score to the score_list
Score_list.append(MLP_result)
algo_names.append("MLP")
```

```python
print("The MAE score for the MLP model above is %d " % (MLP_result))
```

```python
# Now time for us to analyze the performance and accuracy of the models 

# Set figure size
plt.figure(figsize=(25, 10))

#Plot the MAE of all algorithms
fig, ax = plt.subplots()
plt.plot(Score_list)

#Set the tick names to names of combinations
ax.set_xticks(range(len(algo_names)))
ax.set_xticklabels(algo_names,rotation='vertical')

#Plot the accuracy for all combinations
plt.show()  
```
